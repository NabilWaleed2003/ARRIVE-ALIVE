```python
import os
import torch
print(torch.__version__)


data = "Driver Drowsiness Dataset (DDD)"
print(os.listdir(data))
```

```python
import os
import random
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, Dataset
from torchvision import transforms, datasets
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

# تحميل البيانات
Drowsy = [os.path.join(data, 'Drowsy', p) for p in os.listdir(os.path.join(data, 'Drowsy'))]
Non_Drowsy = [os.path.join(data, 'Non Drowsy', p) for p in os.listdir(os.path.join(data, 'Non Drowsy'))]

all_data = Drowsy + Non_Drowsy
random.shuffle(all_data)

train_data = all_data[:int(.8 * len(all_data))]
test_data = all_data[int(.8 * len(all_data)):]

# التحويلات
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor()
])

# تعريف الداتا
class Data(Dataset):
    def __init__(self, data, transform):
        self.data = data
        self.transform = transform
        self.labels = [1 if 'Drowsy' in path else 0 for path in data]

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        img_path = self.data[idx]
        img = Image.open(img_path).convert('RGB')
        img = self.transform(img)
        label = 1 if 'Drowsy' in img_path else 0
        label = torch.tensor(label, dtype=torch.long)
        return img, label

train_dataset = Data(train_data, transform)
train_dataloader = DataLoader(train_dataset, batch_size=32, shuffle=True)

# دالة لحساب المتوسط والانحراف المعياري
def calculate_mean_std(dataloader):
    mean = torch.zeros(3)
    std = torch.zeros(3)
    total_images = 0

    for images, _ in dataloader:
        batch_size, channels, height, width = images.shape
        total_images += batch_size
        mean += images.mean([0, 2, 3]) * batch_size
        std += images.std([0, 2, 3]) * batch_size

    mean /= total_images
    std /= total_images
    return mean, std

# حساب القيم
mean, std = calculate_mean_std(train_dataloader)
print(f"Mean: {mean.tolist()}")
print(f"Std: {std.tolist()}")


```

```python

# تحديث التحويلات مع القيم المحسوبة
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=mean.tolist(), std=std.tolist())
])

# إعادة تحميل البيانات بالتحويلات الجديدة
train_dataset = Data(train_data, transform)
test_dataset = Data(test_data, transform)

train_dataloader = DataLoader(train_dataset, batch_size=32, shuffle=True)
test_dataloader = DataLoader(test_dataset, batch_size=32, shuffle=False)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim  
import torch.nn.functional as F

class CNN(nn.Module):
    def __init__(self):
        super(CNN, self).__init__()
        self.conv1 = nn.Conv2d(3, 16, kernel_size=3, stride=1, padding=1)
        self.bn1 = nn.BatchNorm2d(16)
        self.conv2 = nn.Conv2d(16, 32, kernel_size=3, stride=1, padding=1)
        self.bn2 = nn.BatchNorm2d(32)
        self.conv3 = nn.Conv2d(32, 64, kernel_size=3, stride=1, padding=1)
        self.bn3 = nn.BatchNorm2d(64)
        self.conv4 = nn.Conv2d(64, 128, kernel_size=3, stride=1, padding=1)
        self.bn4 = nn.BatchNorm2d(128)

        self.relu = nn.ReLU()
        self.pool = nn.MaxPool2d(kernel_size=2, stride=2)
        self.dropout = nn.Dropout(0.3)

        self.flatten = nn.Flatten()
        
        self.fc1 = nn.Linear(128 * 14 * 14, 128)
        self.fc2 = nn.Linear(128, 2)

    def forward(self, x):
        x = self.pool(self.relu(self.bn1(self.conv1(x))))
        x = self.pool(self.relu(self.bn2(self.conv2(x))))
        x = self.pool(self.relu(self.bn3(self.conv3(x))))
        x = self.pool(self.relu(self.bn4(self.conv4(x))))

        x = self.flatten(x)  # يحوله إلى (batch, 128)
        
        x = self.dropout(F.relu(self.fc1(x)))
        x = self.fc2(x)  
        
        return x
    #     self.conv1 = nn.Conv2d(3, 16, kernel_size=3, stride=1, padding=1)
    #     self.bn1 = nn.BatchNorm2d(16)
    #     self.conv2 = nn.Conv2d(16, 32, kernel_size=3, stride=1, padding=1)
    #     self.bn2 = nn.BatchNorm2d(32)
    #     self.conv3 = nn.Conv2d(32, 64, kernel_size=3, stride=1, padding=1)
    #     self.bn3 = nn.BatchNorm2d(64)
    #     self.conv4 = nn.Conv2d(64, 128, kernel_size=3, stride=1, padding=1)
    #     self.bn4 = nn.BatchNorm2d(128)
    #     self.relu = nn.ReLU()
    #     self.pool = nn.MaxPool2d(kernel_size=2, stride=2)
    #     self.dropout = nn.Dropout(0.5)
    #     self.fc1 = nn.Linear(64 * 28 * 28, 512)
    #     self.fc2 = nn.Linear(512, 2)  
        
    # def forward(self, x):
    #     x = self.pool(self.relu(self.conv1(x)))
    #     x = self.pool(self.relu(self.conv2(x)))
    #     x = self.pool(self.relu(self.conv3(x)))

    #     x = x.view(x.size(0), -1) 
    #     x = self.relu(self.fc1(x))
    #     x = self.dropout(x)
    #     x = self.fc2(x)
    #     print(x.shape)
    #     return x
model = CNN()
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = model.to(device)
torch.manual_seed(42)
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
print(model)
```

```python
losses = []
num_epochs = 5


for epoch in range(num_epochs):
    model.train()
    total_loss = 0.0
    for images, labels in train_dataloader:
        images, labels = images.to(device), labels.to(device)
        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        total_loss += loss.item()
        avg_loss = total_loss / len(train_dataloader)
    losses.append(avg_loss)
    print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {avg_loss:.4f}')


    # losses.append(total_loss)
    # print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {total_loss:.4f}')
        
```

```python
model.eval()
correct = 0
total = 0
with torch.no_grad():
  for images, labels in test_dataloader:
    images, labels = images.to(device), labels.to(device)
    outputs = model(images)
    _, predicted = torch.max(outputs.data, 1)
    total += labels.size(0)
    correct += (predicted == labels).sum().item()

accuracy = 100 * correct / total
print(f'Test Accuracy: {accuracy:.2f}%')
```

```python
from fastapi import FastAPI, File, UploadFile
from fastapi.responses import JSONResponse
import torch
import torchvision.transforms as transforms
from PIL import Image
import io
import nest_asyncio

nest_asyncio.apply()



app = FastAPI()


model = torch.jit.load("model.path")
model.eval()


transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=mean.tolist(), std=std.tolist())
])

# دالة التنبؤ
def predict_image(image):
    image = transform(image).unsqueeze(0)  
    with torch.no_grad():
        output = model(image)
    prediction = torch.argmax(output, dim=1).item() 
    return "Drowsy" if prediction == 1 else "Non-Drowsy"

# إنشاء API لاستقبال الصور
@app.post("/predict/")
async def predict(file: UploadFile = File()):
    try:
        # قراءة الصورة وتحويلها إلى PIL
        image = Image.open(io.BytesIO(await file.read())).convert("RGB")
        prediction = predict_image(image)
        return JSONResponse(content={"prediction": prediction})
    except Exception as e:
        return JSONResponse(content={"error": str(e)}, status_code=500)

# تشغيل السيرفر
if __name__ == "__main__":
    import uvicorn
    uvicorn.run(app, host="0.0.0.0", port=8000)

```
